Project Questions (Matplotlib)

Q1. Analyze Monthly Sales Trend using Line Plot
Objective

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/ecommerce_orders_dataset.csv')

monthly_sales = df.groupby("Month")["Order_Amount"].sum()

plt.figure(figsize=(10,5))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```

Q2. Compare Sales Across Product Categories using Bar Chart
Objective


```python
category_sales = df.groupby("Product_Category")["Order_Amount"].sum()

plt.figure(figsize=(10,5))
plt.bar(category_sales.index, category_sales.values)
plt.xticks(rotation=45)
plt.title("Sales by Product Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

Q3. Visualize Customer Age Distribution using Histogram
Objective

```python
plt.figure(figsize=(8,5))
plt.hist(df["Customer_Age"], bins=20)
plt.title("Customer Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

Q4. Analyze Payment Method Share using Pie Chart
Objective

```python
payment = df["Payment_Method"].value_counts()

plt.figure(figsize=(7,7))
plt.pie(payment,
        labels=payment.index,
        autopct="%1.1f%%")
plt.title("Payment Method Distribution")
plt.show()
```

Q5. Compare Profit Distribution by Membership Status using Box Plot
Objective

```python
gold = df[df["Membership_Status"]=="Gold"]["Profit_Amount"]
silver = df[df["Membership_Status"]=="Silver"]["Profit_Amount"]
bronze = df[df["Membership_Status"]=="Bronze"]["Profit_Amount"]

plt.boxplot([gold,silver,bronze],
            labels=["Gold","Silver","Bronze"])
plt.title("Profit Distribution by Membership")
plt.show()
```

Q6. Relationship Between Order Amount and Profit Amount using Scatter Plot

```python
plt.scatter(df["Order_Amount"],
            df["Profit_Amount"])

plt.xlabel("Order Amount")
plt.ylabel("Profit Amount")
plt.title("Sales vs Profit")
plt.show()
```

Q7. Top 10 Brands by Revenue
Objective

```python
top_brands = df.groupby("Brand")["Order_Amount"]\
               .sum()\
               .sort_values(ascending=False)\
               .head(10)

plt.figure(figsize=(10,5))
plt.bar(top_brands.index,
        top_brands.values)

plt.xticks(rotation=45)
plt.title("Top 10 Brands by Sales")
plt.show()
```

Q8. Returned vs Non-Returned Orders
Objective

```python
returns = df["Returned"].value_counts()

plt.bar(returns.index,
        returns.values)

plt.title("Returned Orders Analysis")
plt.xlabel("Returned")
plt.ylabel("Count")
plt.show()
```

Question 9
Compare average Review Rating by Shipping Method.

```python
rating=df.groupby("Shipping_Method")["Review_Rating"].mean()

plt.bar(rating.index,rating.values)
plt.xticks(rotation=45)
plt.title("Review Rating")
plt.show()
```

Question 10
Plot Profit Margin Percentage Distribution.

```python
plt.hist(df["Profit_Margin_Percent"],bins=20)
plt.title("Profit Margin Distribution")
plt.xlabel("Profit %")
plt.show()
```

Q6. How do Review Ratings differ among Product Categories?
Objective

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df,
            x='Product_Category',
            y='Review_Rating')
plt.xticks(rotation=45)
plt.title("Ratings by Product Category")
plt.show()
```

Q7. Product Category vs Gender Sales

```python
sns.barplot(data=df,
            x="Product_Category",
            y="Order_Amount",
            hue="Customer_Gender")

plt.xticks(rotation=45)
plt.show()
```

Q8. Which Country generates the highest sales?
Objective

```python
country_sales = df.groupby('Country')['Order_Amount'].sum().reset_index()

plt.figure(figsize=(10,5))
sns.barplot(data=country_sales,
            x='Country',
            y='Order_Amount')
plt.xticks(rotation=45)
plt.title("Country-wise Sales")
plt.show()
```

Q9. What is the correlation among numerical variables?
Objective

```python
corr = df[['Order_Amount',
           'Profit_Amount',
           'Customer_Age',
           'Quantity',
           'Review_Rating']].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr,
            annot=True,
            cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

Q10. Does Customer Gender influence Order Amount?
Objective

```python
plt.figure(figsize=(8,5))
sns.violinplot(data=df,
               x='Customer_Gender',
               y='Order_Amount')
plt.title("Order Amount by Gender")
plt.show()
```
