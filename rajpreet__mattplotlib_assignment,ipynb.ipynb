```python
import matplotlib.pyplot as plt
import numpy as np
products = ["laptop","phone","tablet"]
sales_2024 = [100,150,80]
sales_2025 = [120,170,100]
x = np.arange(len(products))
width = 0.35
plt.bar(x-width/2,sales_2024,width,label="2024")
plt.bar(x+width/2,sales_2025,width,label="2025")
plt,plt.xticks(x,products)
plt.title("sales comparison")
plt.xlabel("products")
plt.ylabel("sales")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
products = ["laptop","phone","tablet"]
sales_2024 = [100,150,80]
sales_2025 = [120,170,100]
x = np.arange(len(products))
width = 0.35
plt.bar(
    x-width/2,
    sales_2024,
    width/2,
    color="skyblue",
    label="2024"
)
plt.bar(
    x+width/2,
    sales_2025,
    width/2,
    color="orange",
    label="2025"

)
plt.xticks(x,products)
plt.title("sales comparison")
plt.xlabel("products")
plt.ylabel("sales")
plt.legend()
plt.show()
```

```python
subjects = ["math","science","english"]
marks_2024 = [80,75,90]
marks_2025 = [85,80,95]
x = np.arange(len(subjects))
width = 0.35
plt.bar(
    x-width/2,
    marks_2024,
    width,
    color="red",
    label="2024"
)
plt.bar(
    x+width/2,
    marks_2025,
    width,
    color ="green",
    label="2025"
)
plt.xticks(x,subjects)
plt.title("marks comparison")
plt.xlabel("subjects")
plt.ylabel("marks")
plt.legend()
plt.show()
```

```python
products = ["laptop","phone","tablet"]
online = [50,70,40]
offline = [30,40,20]
plt.bar(
    products,
    online,
    label="online"
)
plt.bar(
    products,
    offline,
    label="offline"
)
plt.title("sales comparison")
plt.ylabel("marks")
plt.legend()
plt.show()
```

```python
students= ["aman","riya","john"]
theory = [50,70,40]
practical = [30,40,20]
plt.bar(
    students,
    theory,
    label="theory"
)
plt.bar(
    students,
    practical,
    bottom=theory,
    label="practical"
)
plt.title("student marks")
plt.ylabel("marks")
plt.legend()
plt.show()
```
